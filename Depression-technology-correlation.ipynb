# filtering fourth data set

```python
Suicide_rates.head(1) # The data headers are viewed so relevent data can be selected.
```

```python
Suicide_rates = Suicide_rates[["country", "year", "suicides_no"]] # Relevent columns are selected from the dataframe.
```

```python
Suicide_rates.head(1) # The data headers are viewed.
```

```python
Suicide_rates = Suicide_rates[(Suicide_rates['year'] >= 2002) & (Suicide_rates['year'] <= 2019)] # Data from required year 
                                                                                        # is selected form the dataframe.
```

```python
Suicide_rates # The data frame is viewed.
```

```python
# Creating a filter to remove extra countries so the data is consistaent below. 
datafilter = Suicide_rates.groupby(["country"],as_index=False).nunique() #  Grouping counries with years as 
                                                                        # how many times a year appears in a countr.y
datafilter = datafilter[(datafilter["year"]==14)] # selecting only countries with years appearing 14 times.
datafilter.head(3) # the data headers are viewed.
```

```python
datafilter = datafilter[["country"]] # The relevent column is selected from the dataframe.
datafilter.head(3) # the data headers are viewed .
```

```python
Suicide_rates = Suicide_rates[Suicide_rates.country.isin(datafilter.country)] # only countires/data required is selected.
```

```python
# Below the data is grouped by the column year and the data form the column Sucide_no is added together below.
Suicide_rates = Suicide_rates.groupby(['year'],as_index=False).agg({'suicides_no': sum})
```

```python
Suicide_rates # the  dataframe is viewed.
```

```python
# Below the second column name is changed.
Suicide_rates = Suicide_rates.rename(columns={'suicides_no': 'suicide rate', 'year': 'Year'})
```

```python
Suicide_rates # the dataframe is viewed.
```

```python
SocialMediaUsageByYear # The dataframe is viewed.
```

```python
# Data from two dataframes is merged together based on common years.
SucideRatesAndSocialMediaUsageByYear = pd.merge(SocialMediaUsageByYear, Suicide_rates, on="Year")
```
